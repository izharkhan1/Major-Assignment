```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('imdb_movies.csv')
movies_data["index"] = movies_data.index
movies_data.head()
```

```python
null_values = movies_data.isnull().sum()
null_values
```

```python
selected_column = ['names','genre','country','orig_lang']
selected_column
```

```python
for null in selected_column:
    movies_data[null] = movies_data[null].fillna(' ')
```

```python
combined_column = movies_data['names']+''+movies_data['genre']+''+movies_data['country']+''+movies_data['orig_lang']
combined_column
```

```python
vectorizer = CountVectorizer()
feature_extraction = vectorizer.fit_transform(combined_column)
print(feature_extraction)
```

```python
similarity = cosine_similarity(feature_extraction)
print(similarity)
```

```python
movies_name = input("Enter Movie Name : ")
```

```python
all_titles = movies_data['names'].tolist()
all_titles
```

```python
find_close_match = difflib.get_close_matches(movies_name,all_titles)
find_close_match
```

```python
close_match = find_close_match[0]
close_match
```

```python
index_of_movie = movies_data[movies_data.orig_title==close_match]['index'].values[0]
index_of_movie
```

```python
similarity_score = list(enumerate(similarity[index_of_movie]))
similarity_score
```

```python
sorted_similar_movies = sorted(similarity_score,key = lambda x:x[1],reverse = True)
sorted_similar_movies
```

```python
movie_name = input("Enter Your Movie Name : ")

list_of_all_titles = movies_data['names'].tolist()

find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)

close_match = find_close_match[0]

index_of_movie = movies_data[movies_data.orig_title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)





print("Movies suggested for you :  ")
i = 0

for movie in sorted_similar_movies:
    index = movie[0]
    movie_from_index = movies_data[movies_data.index==index]['orig_title'].values[0]
    if i<30:
        print(i,".", movie_from_index)
        i+=1
```
